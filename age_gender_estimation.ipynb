```
import numpy as np
import data_utils
import keras
import config as cf
import matplotlib.pyplot as plt
import cv2
from model import Efficient_Net
```

```
model = Efficient_Net(trainable=True)
```

```
# Load weights
model = Efficient_Net(trainable=False).model
model.load_weights('weight.h5')
```

```
testX, ageTestY, genderTestY = data_utils.data_test()
testX = np.array(testX)
print('x_test shape: ', testX.shape)
ageTestY = np.array(ageTestY)
print('ageTestY shape: ', ageTestY.shape)
genderTestY = np.array(genderTestY)
print('genderTestY shape: ', genderTestY.shape)
```

```
loss, age_output_loss, gender_output_loss, age_output_acc, gender_output_acc = model.evaluate(x=testX, y={'age_output': ageTestY, 'gender_output': genderTestY})
```

```
model.summary()
```

```
print('Result: ')
print('Loss: ', loss)
print('Age output loss: ', age_output_loss)
print('Gender output loss: ', gender_output_loss)
print('Age output accuracy: ', age_output_acc * 100)
print('Gender output accuracy: ', gender_output_acc * 100)
```
